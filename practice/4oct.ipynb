## GMM

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(seed = 42)
```

```python
n = 1000000
component = rng.choice([0,1,2],p = [0.2,0.3,0.5] , size = n)
```

```python
component[:50]
```

```python
mu = [-10 , 0 , 10]
```

```python
x = np.zeros(n)
```

```python
for i , c in enumerate(component):
    x[i] = rng.normal(mu[c],1)
print(x.shape)
```

```python
plt.hist(x , bins = 1000 , edgecolor = 'green')
plt.axvline(mu[0])
plt.axvline(mu[1])
plt.axvline(mu[2])
plt.show()
```

## Histogram

    bins : 
    The bins parameter controls how many intervals (bars) the data is split into, or you can specify the exact bin edges.

```python
data = rng.normal(loc=0, scale=1, size=1000)
```

```python
plt.hist(data, bins=20)  # 20 equal-width bins
plt.show()
```

```python
plt.hist(data, bins=[0, 10, 20, 30, 40, 50])  # bins with specified edges
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(10, 2, 100)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
mean = np.mean(data)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Mean')
plt.show()
```
